```python
np.random.seed(80)
ran_array_with_seed = np.random.randint(1,101,(4,4))
ran_array_col_sum = np.sum(ran_array_with_seed,axis=0)
ran_array_row_sum = np.sum(ran_array_with_seed,axis=1)
ran_array_col_max = np.max(ran_array_with_seed,axis=0)
ran_array_row_max = np.max(ran_array_with_seed,axis=1)
ran_array_col_min = np.min(ran_array_with_seed,axis=0)
ran_array_row_min = np.min(ran_array_with_seed,axis=1)
ran_array_col_avg = np.mean(ran_array_with_seed,axis=0)
ran_array_row_avg = np.mean(ran_array_with_seed,axis=1)

print("Random Array with Seed:\n", ran_array_with_seed,"\n")
print("Column Sum: ", ran_array_col_sum,"\n")
print("Row Sum: ", ran_array_row_sum,"\n")
print("Column Max: " ,ran_array_col_max,"\n")
print("Row Max: ", ran_array_row_max,"\n") 
print("Column Min :", ran_array_col_min,"\n")
print("Row Min :", ran_array_row_min,"\n")
print("Col Avg :", ran_array_col_avg,"\n")
print("Row Avg :", ran_array_row_avg,"\n")
print("Bigger Than 50: ",ran_array_with_seed[ran_array_with_seed>50])
```

# Median - 

## The term "Median" reffers to the value in the middle of a list of numbers. In an odd list it would be equal to the average.

# EXAMPLE: 

## In the array: 
                 [[48 61 39 84]
                  [11 51  4 87]
                  [24 16 82  6]
                  [65 10 93 43]] 
The Median Number is 45.5

```python
median_for_array = np.median(ran_array_with_seed)
print(median_for_array)
```

# standard deviation - 

## The term "standard deviation" reffers to the deviation of data from the mean(average). It shows the deviation of values across a bulk of data.

# Example:

## In this array:
             [[48 61 39 84]
              [11 51  4 87]
              [24 16 82  6]
              [65 10 93 43]] 
The Standard Deviation is: 30.169313880166385

```python
std_for_array = np.std(ran_array_with_seed)
print(std_for_array)
```

# Variance -

## Variance is a measure of how much the values differ from the average

# Example:
## In this array:
         [[48 61 39 84]
          [11 51  4 87]
          [24 16 82  6]
          [65 10 93 43]] 
The variance is: 910.1875

```python
var_for_array = np.var(ran_array_with_seed)
print(var_for_array)
```
